# **1. Introduction to Matplotlib**

**Matplotlib** is a Python library that allows you to create static, animated, and interactive visualizations. Matlab aur Excel ki tarah, lekin Python mein aur bohot zyada customizable.

Matplotlib has two main APIs (ways to make plots):

**A. Pyplot API**

pyplot is like a simple interface to make plots quickly.

You usually import it as:

```python
#!pip install matplot
#!pip install seaborn

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x,y)
plt.title("Simple Line plot")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

`plt.plot` se hum line plot bana sakte hain. `plt.show()` se figure screen pe aa jata hai.

**B. Object-Oriented (OO) API**

- OO API gives more flexibility and control over the plot.

- Instead of using plt, you create figure and axes objects.

```python
#create figure axes 
fig, ax = plt.subplots()  # Create figure and axes
ax.plot(x, y)
ax.set_title("Line Plot with OO API")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
plt.show()
```

Yahaan **ax** object se hum plot ke har ek part ko individually control kar sakte hain.

#### **1. Axes**

Axes is the part of the figure where the actual data is plotted (lines, bars, points, etc.).

A Figure can have one or more Axes, and each Axes can have its own title, labels, and plots.

Think of Figure as a canvas and Axes as the area on that canvas where you draw.

Example:

```python


fig = plt.figure()      # Create a canvas
ax = fig.add_axes([0, 0, 0.5, 0.5])  # Add axes: [left, bottom, width, height]
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_title("Axes Example")
plt.show()
```

**2. Subplot**<br>
Subplot is a way to divide the figure into multiple smaller axes in a grid format.

Useful when you want multiple plots in a single figure.

Syntax: 
> plt.subplot(rows, columns, index)

```python
fig = plt.figure()     # Create a canvas
ax1 = fig.add_axes([0, 0, 1, 1])  # Main axes
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("First Plot")
plt.subplot(1,2,2)
plt.plot(['A','B','C'],[4,1,6])
plt.title("Second Plot")

plt.show()
```

#### Markers <br>
 In Matplotlib markers are used to highlight individual
data points on a plot.

```python
x = [22,3,2,6,4,44,6,43,56,45,64]
y = [34,3,23,43,23,45,54,23,42,53,2]
plt.scatter(x,y,marker='p')

plt.title("Scatter Plot")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

**Matplotlib - Styles** : <br> 
In Matplotlib library styles are configurations that allow
us to change the visual appearance of our plots easily

```python
plt.style.use("ggplot")
x = [22,3,2,6,4,44,6,43,56,45,64]
y = [34,3,23,43,23,45,54,23,42,53,2]
plt.scatter(x,y,marker='p')

plt.title("Scatter Plot")
plt.xlabel("Input")
plt.ylabel("Output")
plt.show()
```

**Matplotlib- Legends** <br>
It is a box containing a symbol and a label for each
series in the graph

```python
x = ['A','B','C']
y1 = [1,2,3]
y2 = [4,6,7]
y3 = [6,4,5]

#ploting the data 
line1 = plt.plot(x,y1)
line2 = plt.plot(x,y2)
line3 = plt.plot(x,y3)

#calling legend 
plt.legend([line1,line2,line3],['Label 1','Label 2','Label 3'])

plt.show()
```

```python
import numpy as np
x = np.linspace(0 , 2*np.pi , 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize = (7,4))

ax.grid(True)

ax.set_title("Basic plot with grid")

plt.xlim(1,5)

plt.plot(x,y)
plt.show()
```

**Matplotlib- Ticks and Tick Labels** <br>
In the context of graphs and plotting, ticks are the
small lines that show the scale of x and y-axes,
providing a clear representation of the value
associated with each tick.

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)


plt.plot(x,y)
plt.title("Sine wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")




plt.show()
```
